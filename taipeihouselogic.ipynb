```python
df = df.drop(['Unnamed: 8','Unnamed: 9'], axis=1)
df
```

```python
df = df.drop(['Unnamed: 10','Unnamed: 11','Unnamed: 12','Unnamed: 13'], axis=1)
df
```

```python
df.down = df.down.map({'已拆遷':0, '列管':1, '解除列管':2})
df
```

```python
df.zone = df.zone.map({'信義':0,'北投':1, '士林':2,'大安':3,'中山':4,'內湖':5,'南港':6,'文山':7})
df
```

```python
df = df.drop(['Demolitiontime'], axis=1)
df
```

```python
df.Location = df.no
df
```

```python
from sklearn.model_selection import train_test_split

y = df.down
X = df.drop('down', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()

clf.fit(X_train, y_train)
```

```python
clf.score(X_train, y_train)
```
